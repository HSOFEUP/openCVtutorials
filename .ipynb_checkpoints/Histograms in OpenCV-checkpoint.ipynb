# Visualize Individual Color Components

```python
import cv2
import numpy as np

from matplotlib import pyplot as plt # import matplotlib to create histograms 
```

```python
input = cv2.imread('C:\\Users\\Zahid\\Desktop\\Deep Learning\\OpenCV Tutorials\\input.jpg')

histogram = cv2.calcHist([input] , [0] , None , [256] , [0,256] ) #histogram of blue color of the image
histogram_1 = cv2.calcHist([input] , [1] , None , [256] , [0,256] ) #histogram of Greens in image
histogram_2 = cv2.calcHist([input] , [2] , None , [256] , [0,256] ) #histogram of reds

''' Parameters of Histogram
first input is the image , it is in square brackets since it needs to be in an array form
second parameter is the number of channels to get 0 , 1 or 2 to get colors such as red green or blue.
third parameter for mask
fourth parameter is the histogram size
fifth parameter is the range of the histogram
'''
#ravel() is used to convert 2d array to 1d array

plt.hist(input.ravel() , 256 , [0,256])
plt.show()

plt.hist(histogram , 256 , [0,256])
plt.show()

plt.hist(histogram_1 , 256 , [0,256])
plt.show()

plt.hist(histogram_2 , 256 , [0,256])
plt.show()

color = ('b' , 'g' , 'r')
#sepearting different color channels

#seperating the colors and plotting each for the histogram

for i, col in enumerate(color):
    histogram2 = cv2.calcHist([input] , [i] , None , [256] , [0,256])
    plt.plot(histogram2 , color = col )
    plt.xlim([0,256])
    
plt.show()
```
